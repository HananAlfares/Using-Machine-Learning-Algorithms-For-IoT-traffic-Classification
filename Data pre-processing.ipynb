```python
# make the labels clear 
df.loc[(df.label == '-   Malicious   PartOfAHorizontalPortScan'), 'label'] = 'PartOfAHorizontalPortScan'
df.loc[(df.label == '(empty)   Malicious   PartOfAHorizontalPortScan'), 'label'] = 'PartOfAHorizontalPortScan'
df.loc[(df.label == '-   Malicious   Okiru'), 'label'] = 'Okiru'
df.loc[(df.label == '(empty)   Malicious   Okiru'), 'label'] = 'Okiru'
df.loc[(df.label == '-   Benign   -'), 'label'] = 'Benign'
df.loc[(df.label == '(empty)   Benign   -'), 'label'] = 'Benign'
df.loc[(df.label == '-   Malicious   DDoS'), 'label'] = 'DDoS'
df.loc[(df.label == '-   Malicious   C&C'), 'label'] = 'C&C'
df.loc[(df.label == '(empty)   Malicious   C&C'), 'label'] = 'C&C'
df.loc[(df.label == '-   Malicious   Attack'), 'label'] = 'Attack'
df.loc[(df.label == '(empty)   Malicious   Attack'), 'label'] = 'Attack'
df.loc[(df.label == '-   Malicious   C&C-HeartBeat'), 'label'] = 'C&C-HeartBeat'
df.loc[(df.label == '(empty)   Malicious   C&C-HeartBeat'), 'label'] = 'C&C-HeartBeat'
df.loc[(df.label == '-   Malicious   C&C-FileDownload'), 'label'] = 'C&C-FileDownload'
df.loc[(df.label == '-   Malicious   C&C-Torii'), 'label'] = 'C&C-Torii'
df.loc[(df.label == '-   Malicious   C&C-HeartBeat-FileDownload'), 'label'] = 'C&C-HeartBeat-FileDownload'
df.loc[(df.label == '-   Malicious   FileDownload'), 'label'] = 'FileDownload'
df.loc[(df.label == '-   Malicious   C&C-Mirai'), 'label'] = 'C&C-Mirai'
df.loc[(df.label == '-   Malicious   Okiru-Attack'), 'label'] = 'Okiru-Attack'
df['label'].value_counts()
```

```python
# splitting the data into X,y
X=df.iloc[:,:-1]
y=df.iloc[:,-1]

```

```python
# drop the silent features
X= X.drop(columns=['ts','uid','id.orig_h','id.orig_p','id.resp_h','id.resp_p',
                               'service','local_orig','local_resp','history'])
```

```python
# fill the missing values with zeros 
X.fillna(0,inplace=True)
```

```python
# check duplicated data type
# we will keep them because they outputed from different secnarios 
print(df.duplicated)
```

```python

# to deal with the duration , conn_state, proto by converting the (-) into numerical(0)
X['proto'].value_counts()
X['proto'] = X['proto'].str.replace('0','null')

X["conn_state"]=X["conn_state"].str.replace("-","0")
X['duration'] = X['duration'].str.replace('-','0')
# to deal with the bytes 
X['orig_bytes'] = X['orig_bytes'].str.replace('-','0')
X['resp_bytes'] = X['resp_bytes'].str.replace('-','0')
```

```python
# label encoder of categorical data 
Unique=X.iloc[:,:].nunique() #this important for label encoder
labelencoder=LabelEncoder() #Object defined
X["proto"]=labelencoder.fit_transform(X["proto"])
X["conn_state"]=labelencoder.fit_transform(X["conn_state"])
X["orig_bytes"]=labelencoder.fit_transform(X['orig_bytes'])
X["resp_bytes"]=labelencoder.fit_transform(X['resp_bytes'])
X["duration"] =labelencoder.fit_transform(X["duration"])
```

```python
# convert the categorical data of y using label encoder  
y=labelencoder.fit_transform(y)
```

```python
#splitting the data into test and train set :    
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
```

```python
# perform feature scailing using Standerd scaler 
st= StandardScaler()
st.fit_transform(X_train)
st.fit(X_test)
```
